```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = ('/content/swiss_office_load.csv')
data = pd.read_csv(csv_path, sep=';') 
```

```python
df = pd.read_csv(csv_path, header=None,sep=';', names=['Seconds', 'Cooling', 'Heating', 'Water'])
```

```python
df
```

```python
# Plot the space heating and cooling loads for every second in a year
fig, ax = plt.subplots()
ax.plot(df['Seconds'], df['Cooling'], label='Cooling')
ax.plot(df['Seconds'], df['Heating'], label='Heating')
ax.set_xlabel('Seconds in the Year')
ax.set_ylabel('Load (Watts)')
ax.set_title('SwissOffice_20190916 Space Heating and Cooling Loads')
ax.legend()
plt.show()
```
